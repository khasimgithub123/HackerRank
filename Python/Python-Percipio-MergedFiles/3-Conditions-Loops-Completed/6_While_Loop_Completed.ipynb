Convert a list of values from Celsius to fahrenheit<br>
Celsius to fahrenheit = (C  9/5) + 32 = F

```python
c_degrees = [-20, -15, -10, -5, 0, 5, 10, 15, 20, 25, 30, 35, 40]
index = 0

print( '    C    F')

while index < len(c_degrees):
    
    C = c_degrees[index]
    F = (9.0/5)*C + 32
    
    print( '%5d %5.1f' % (C, F))
    
    index += 1
```

In this example, we have a list of vegetable which contains the fruit Orange. We examined from the first element to the end, if we find the element we will print the position of the fruit and delete the fruit from the list and escape the loop.<br>Then print the updated vegetable list

```python
clean_haystack =["hay", "hay", "hay", "hay",
                 "hay", "hay", "hay", "hay"]

unclean_haystack =["hay", "hay", "hay", "hay",
                   "hay", "needle", "hay", "hay"]

searching_for = "needle"
```

```python
maximum = len(clean_haystack)

i = 0

while i < maximum:
    
    if searching_for == clean_haystack[i]:
        print("The %s is at index %i:" % (searching_for, i))
        
        del clean_haystack[i]
        break
        
    i = i + 1

else:
    print("\nThe needle was not found :(")
```

```python
maximum = len(unclean_haystack)

i = 0

while i < maximum:
    
    if searching_for == unclean_haystack[i]:
        print("The %s is at index %i :)" % (searching_for, i))
        
        del unclean_haystack[i]
        print ('The updated haystack is:', unclean_haystack)
        break
        
    i = i + 1

else:
    print("\nThe needle was not found :(")
```

Printing all the prime numbers between 2 and the given value using break statement

```python
max_num = int(input('enter an integer: '))
num = 2

print("prime numbers upto %i: "% max_num)

while num < max_num + 1: 
    
    isPrime = True
    count = 2
    
    while count < num:
        if num % count == 0:
            isPrime = False
            break
            
        count += 1

    if isPrime:
        print(num)

    num += 1
```

#### Use of pass in while loop<br>
When an external condition is triggered, the pass statement allows you to handle the condition without the loop being impacted in any way; all of the code will continue to be read unless a break or other statement occurs.

```python
num_value = 10

while num_value > 0:              
    print ("current value is :", num_value)
    
    if num_value == 5:
        pass
        print ('this is a pass block')
        
    num_value = num_value - 1

print ("good bye!")
```

The continue statement rejects all the remaining statements in the current iteration of the loop and moves the control back to the top of the loop.<br>
Using continue statement you can skip elements in your program


Here i am printing all the odd numbers between 0 to 10 and skipping even numbers using continue

```python
num = 0                    

while num < 10:          
    
    num = num + 1
    
    if num % 2 == 0:
        continue
        
    print ("odd number :", num) 
    
print ("good bye!")
```

Here i am adding all numbers between 0 to 50 in my list<br>
Except those numbers which is divisible by 5 and 7 using continue statement

```python
my_list = []
num = 0

while num in range(0, 50):
    
    num += 1
    
    if num % 5 == 0 or num % 7 == 0:
        continue
        
    my_list.append(num)
    
    
print(my_list)
```

In the following example, while loop is set to print the first 9 items in the tuple starting from 9 to 1. But when the count is divisible by 4 the continue statement will execute and skip the element.

```python
num_tup = (1, 2, 3, 4, 5, 6, 7, 8, 9)
count = 9

while (count > 0):
    count = count - 1
    
    if num_tup[count] % 4 == 0:
        continue
        
    print (num_tup[count])
    
print ("end of the program")
```

```python
my_str = "pywthcont"
correct_str = "python"


my_index = 0
correct_index = 0

while my_index < len(my_str) and correct_index < len(correct_str):

    if my_str[my_index] != correct_str[correct_index]:
        my_index += 1
        continue
        
    print(my_str[my_index])
    my_index += 1
    correct_index += 1
```

#### We can also use break and continue statement together in the same loop

Here we are getting the cube of a input number<br>
If the input is a number then the program will execute and it will print the result<br>
If the input is `out` then the break statement will be execute and we will come out from the loop<br>
If the input is not a digit then the continue statement will be execute and move to the next iteration

```python
while True:
    
    value = input("\nEnter a number: ")
 
    if value == "out": 
        print("Good bye! (break statement executed)")
        break
 
    if not value.isdigit(): # if input is not a digit move on to the next iteration
        print("Enter integers only (continue statement executed)")
        continue
 
    value = int(value)
    print("Cube of %d is %d" %(value, value ** 3))
```
